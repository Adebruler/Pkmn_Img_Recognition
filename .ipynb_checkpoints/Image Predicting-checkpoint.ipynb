```python
# Dependencies
import matplotlib.pyplot as plt
%matplotlib inline

import os
import numpy as np
import tensorflow as tf
import pandas as pd

import keras
from keras.preprocessing import image
import operator
```

```python
 # Load the model
from keras.models import load_model
sm = load_model("models/sugimachine.h5")
```

```python
types = sorted(["Normal", "Fire", "Water", "Electric", "Grass", "Ice", "Fighting", "Poison", "Ground", "Flying", "Psychic", 
         "Bug", "Rock", "Ghost", "Dragon", "Dark", "Steel", "Fairy"])
print(types)
```

```python
# Read in DataFrame of stored images and types
pokedex_df=pd.read_csv("Data/Pokedex.csv").drop(labels="Unnamed: 0",axis=1)
# Convert "Type" column (read in as strings) back to lists
pokedex_df.Type=pokedex_df.Type.str[1:-1].str.split(', ').tolist()
for i in range(809):
    pokedex_df.type[i] = [t.replace("'","") for t in pkoedex_df.Type[i]]
pokedex_df.head()
```

```python
 # Default Image Size for Sugimachine
image_size = (200, 200)
```

```python
def predict(image_path):
    img = image.load_img(image_path, target_size=image_size)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
#     x = preprocess_input(x)
    predictions = sm.predict(x)
    top2 = best2(predictions[0])
    return top2
```

```python
def best2(predictions):
    predictionary = {}
    for i in range(18):
        predictionary[types[i]]=predictions[i]
    sorted_x = sorted(predictionary.items(), key=operator.itemgetter(1), reverse=True)
    return sorted_x[:2]
```

```python
Predict1 = []
Predict2 = []
Accuracy = []
for mon in range(809):
    montype = [i.replace("'","") for i in pokedex_df["Type"][mon]]
    image_path = os.path.join("Images", pokedex_df["Files"][mon])
    top2 = predict(image_path)
    top2types = [type[0] for type in top2]
    Predict1.append(top2types[0])
    Predict2.append(top2types[1])
    acc = 0
    if len(montype) == 1:
        if montype[0] == top2types[0]:
            acc = 1
        elif montype[0] == top2types[1]:
            acc = 0.5
    else:
        if montype[0] in top2types:
            acc += 0.5
        if montype[1] in top2types:
            acc += 0.5
    print(f"Accuracy {mon}: {acc}")
    Accuracy.append(acc)

predictions_df = pd.DataFrame({"Predict1":Predict1, "Predict2":Predict2, "Accuracy":Accuracy})
```

```python
pokedex_predict_df = pokedex_df.merge(predictions_df,left_index=True,right_index=True)
pokedex_predict_df.head()
```

```python
pokedex_df.to_csv("Data/Pokedex_Predictions.csv")
```
