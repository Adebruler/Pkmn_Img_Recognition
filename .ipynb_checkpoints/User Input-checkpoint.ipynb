```python
# Dependencies
import matplotlib.pyplot as plt
%matplotlib inline

import os
import numpy as np
import tensorflow as tf
import pandas as pd

import keras
from keras.preprocessing import image
import operator
```

```python
 # Load the model
from keras.models import load_model
sm = load_model("models/sugimachine.h5")
```

```python
types = sorted(["Normal", "Fire", "Water", "Electric", "Grass", "Ice", "Fighting", "Poison", "Ground", "Flying", "Psychic", 
         "Bug", "Rock", "Ghost", "Dragon", "Dark", "Steel", "Fairy"])
print(types)
```

```python
 # Default Image Size for Sugimachine
image_size = (200, 200)
```

```python
def predict(image_path, showpic=False):
    img = image.load_img(image_path, target_size=image_size)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    if showpic:
        plt.imshow(img)
    predictions = sm.predict(x)
    top2 = best2(predictions[0])
    return top2
```

```python
def best2(predictions):
    predictionary = {}
    for i in range(18):
        predictionary[types[i]]=predictions[i]
    sorted_x = sorted(predictionary.items(), key=operator.itemgetter(1), reverse=True)
    return sorted_x[:2]
```

```python
image_path = os.path.join("User_Input", "Gary.png")
top2 = predict(image_path, True)
print(top2)
```
