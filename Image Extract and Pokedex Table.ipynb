```python
# Import Dependencies
import pandas as pd
import requests
from bs4 import BeautifulSoup
import urllib
```

```python
# Data Source: BulbaPedia
url = 'https://bulbapedia.bulbagarden.net/wiki/List_of_Pokémon_by_National_Pokédex_number'
```

```python
# Create soup response to find tables
pkmn_tables=[]
while len(pkmn_tables) < 1:
    response=requests.get(url)
    soup=BeautifulSoup(response.text,'html.parser')
    pkmn_tables=soup.find_all('table')[1:8]
```

```python
# merge pkmn tables from all generations
all_pkmn_df=pd.DataFrame({})
# Gather links for all of the Pokemon's individual pages
pkmn_links =[]
for region in pkmn_tables:
#     Find all links in each table
    links = region.find_all('a')
    for link in links:
#       Last 4 characters of link can determine if it is for a pokemon
        currlink = link['href']
        linktype = currlink[-4:]
#       Add only pokemon links to list
        if (linktype == 'mon)' and (currlink not in pkmn_links)):
            pkmn_links.append(currlink)
            print(currlink)
#     trim header row to enable merge
    gen_df = pd.read_html(region.prettify())[0][1:]
    all_pkmn_df=all_pkmn_df.append(gen_df)
# reindex combined table
count=len(all_pkmn_df)
indices=[i for i in range(count)]
all_pkmn_df.index=indices
```

```python
all_pkmn_df
```

```python
baseurl = 'https://bulbapedia.bulbagarden.net'
# Store off links for the main images of each pokemon
img_links = []
# Follow the previous links to each pokemon's page
for pkmn_link in pkmn_links:
    url = baseurl+pkmn_link
#     Test for connection. If no images are found, repeat request until found
    images = []
    while len(images) < 1:
        response=requests.get(url)
        soup=BeautifulSoup(response.text,'html.parser')
        images = soup.find_all('img')
#     The main images are consistently the first image sized to width="250"
    for image in images:
        if (image["width"]=="250"):
            img_link = image["src"]
            print(img_link)
            img_links.append(img_link)
            break
```

```python
# Save images from links to local Images directory
filenames=[]
for img_link in img_links:
    filename=img_link.split('/')[-1][6:]
    filepath = f"Images/{filename}"
    img_data = requests.get(f"https:{img_link}").content
    filenames.append(filename)
    print(filename)
    with open(filepath, 'wb') as handler:
        handler.write(img_data)
```

```python
first_form_df = all_pkmn_df.drop_duplicates(subset=3,keep="first")
# Drop regional pokedex number and picture column
nat_pkmn_df=first_form_df[[1,3,4,5]]
# Remove pound from national pokedex numbers
pkmn_count=len(nat_pkmn_df)
nat_pkmn_df.index = [i for i in range(pkmn_count)]
combined_types = []
for form in range(pkmn_count):
    nat_pkmn_df[1][form]=nat_pkmn_df[1][form][1:]
#     Combine types into one columnof lists
    type1 = nat_pkmn_df[4][form]
    type2 = nat_pkmn_df[5][form]
    combined_type = [type1]
    if type(type2) == type('str'):
        combined_type.append(type2)
    combined_types.append({'Type':combined_type})

types_col = pd.DataFrame(combined_types)
# Rename Columns
nat_pkmn_df=nat_pkmn_df[[1,3]]
nat_pkmn_df=nat_pkmn_df.rename(columns={1:'Nat_Dex',3:'Name'}).merge(types_col,left_index=True,right_index=True)

nat_pkmn_df.head()
```
